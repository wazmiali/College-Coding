```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
# Note: Seaborn has some built in datasets 
```

```python
# # Total bill vs tip dataset 
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Setting a theme for the plots
sns.set_theme()
```

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',            # Fix: changed 'cal' to 'col'
    hue='smoker',
    style='smoker',
    size='size'
)
```

```python
# load the iris dataset 
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# Scatter Plot
```

```python
sns.scatterplot(
    x='sepal_length', 
    y='petal_length', 
    hue='species', 
    data=iris
)
```

```python
# house price dataset

from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load the dataset
house_data = fetch_california_housing()

# Create DataFrame with the name 'house' (no df used)
house = pd.DataFrame(house_data.data, columns=house_data.feature_names)
house['PRICE'] = house_data.target

# Show the first few rows
# print(house.head()) 

house.head()
```

```python
# Distribution of Plot 
```

```python
sns.displot(house['PRICE'])
```

```python
# Correlation 
# Position Correlation
# Negative Correlation
```

```python
# Heat Map
```

```python
correlation = house.corr()
```

```python
# Constructing a heap map 

plt.figure(figsize=(10,10))

sns.heatmap(correlation,cbar=True, square=True, fmt='1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
