```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Toyota.csv', index_col=0, na_values=['??', '????'])
print(data)

# Select numeric columns only
numeric_data = data.select_dtypes(exclude=[object])

# Scatter plot: Age vs. Price
plt.figure(figsize=(8, 6))
plt.scatter(x=numeric_data['Age'], y=numeric_data['Price'], c='blue')
plt.title('Scatter Plot of Age and Price')
plt.xlabel('Age (in months)')
plt.ylabel('Price (in Euros)')
plt.grid(True)
plt.show()
```

```python
# Histogram: Distribution of KM
plt.figure(figsize=(8, 6))
plt.hist(x=numeric_data['KM'], color='green', edgecolor='black', bins=10)
plt.title('Histogram of Kilometers Driven (KM)')
plt.xlabel('Kilometers Driven (KM)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Bar chart: Frequency of Fuel Types
count_list = [979, 120, 100]
fuelType_list = ['Petrol', 'Diesel', 'CNG']
index = np.arange(len(fuelType_list))

plt.figure(figsize=(8, 6))
plt.bar(index, count_list, color=['pink', 'green', 'blue'])
plt.title('Frequency of Different Fuel Types')
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.xticks(index, fuelType_list)
plt.show()
```
